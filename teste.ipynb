```
import urllib
from urllib import request
import re
import pandas as pd

url = "https://www.adorocinema.com/filmes/filme-215094/criticas/espectadores/?page="

def extrai_nome(html):
  inicio_nome = html[n].find('title="') + 7
  fim_nome = html[n].find('">\n<img')
  nome = html[n][inicio_nome:fim_nome]
  return nome

def extrai_infos(html):
  id = html[11:10+11]

  inicio_nome = html.find('title="') + 7
  fim_nome = html.find('">\n<img')
  nome = html[inicio_nome:fim_nome]

  inicio_nota = html.find('"stareval-note">') + 16
  fim_nota = html.find('</span></div>\n<span class="review-card-meta-date')
  nota = html[inicio_nota:fim_nota]

  inicio_data = html.find('date light">\n') + 14 + 10
  fim_data = html.find('\n</span>\n</div>\n</span>\n<div class="content-txt review-card')
  data = html[inicio_data:fim_data]

  inicio_texto = html.find('class="content-txt review-card-content">\n') + len('class="content-txt review-card-content">\n')
  fim_texto = html.find('\n</div>\n<div class="review-card-social"')
  texto = html[inicio_texto:fim_texto]
  
  return id, nome, nota, data, texto

# existem 12 páginas com comentários
qtd_pagina = 12
url_pagina = [url+str(x) for x in range(1, qtd_pagina+1)]
#print(url_pagina)

paginas_raw=[]
for x in url_pagina:
  try:
    response = request.urlopen(x)
    print("URL:",x,"OK")
    raw = response.read().decode('utf8')
    paginas_raw.append(raw)
  except urllib.error.HTTPError as e:
    print("Código erro: %d\nRazão do erro: %s" %(e.code, e.reason))

tabela = []



for p in range(11):

  raw_reviews = []

  encontrar = 'id="review_'
  res = [i.start() for i in re.finditer(encontrar, paginas_raw[p])]
  print(res)
  #print(paginas_raw[0][res[0]:res[1]])


  for i in range(10):
      if i == 9:
          fim = -1
      else:
          fim = res[i+1]
      inicio = res[i]
      raw_review = paginas_raw[p][inicio:fim]
      raw_reviews.append(raw_review)


  for i in range(10):
    id, nome, nota, data, texto = extrai_infos(raw_reviews[i])
    dict = {'id': id, 'nome': nome, 'nota': nota, 'data': data, 'texto': texto}
    tabela.append(dict)
    #print(i)


df = pd.DataFrame(tabela)






```
