```
import nltk
from nltk import word_tokenize
import string
nltk.download('punkt')
nltk.download('mac_morpho')
nltk.download('treebank')
nltk.download('universal_tagset')
```

```
import re
```

```
macm_words = nltk.corpus.mac_morpho.words()
macm_tagd_words=nltk.corpus.mac_morpho.tagged_words()

macm_tagd_sents = nltk.corpus.mac_morpho.tagged_sents()
macm_sents = nltk.corpus.mac_morpho.sents()


# define tamanho de treino e teste

size_sts_tr = int(len(macm_tagd_sents)*0.9)
size_sts_tst = int(len(macm_tagd_sents)*0.1)

#size_w_tr = int(len(macm_tagd_words)*0.9)
#size_w_tst = int(len(macm_tagd_words)*0.1)
```

```
sents_treino = macm_tagd_sents[:size_sts_tr]
sents_teste = macm_tagd_sents[size_sts_tr:]
```

```
#words_treino = macm_tagd_words[:size_w_tr]
#words_teste = macm_tagd_words[size_w_tst:]
```

```
#tags = [tag for (word, tag) in words_treino]
#freq_tag = nltk.FreqDist(tags).max()
#print(freq_tag)
freq_tag = 'N'
```

```
tagger_dft = nltk.DefaultTagger(freq_tag)
print(tagger_dft.evaluate(sents_treino))
```

```
# tagger com expressoes regulares

regex = [
      (r'.*ndo$', 'V'), # gerundio 
      (r'.*oso$', 'ADJ'), # bondoso, perigoso, gostoso 
      (r'.*osa$', 'ADJ'), 
      (r'.*mente$', 'ADV'), # certamente, rapidamente, felizmente  
      #(r'.*mento$', 'N'), # prolongamento, acionamento, afastamento
      (r'.*ção$', 'N'), # poluição, evolução, comunicação, aviação
      (r'.*ções$', 'N'), 
      #(r'.*eza$', 'N'),  # clareza, lerdeza, pobreza
      #(r'.*são$', 'N'), # alusão, previsão, exclusão
      (r'.*ade$', 'N'), # bondade, dificuldade, lealdade
      #(r'.*cio$', 'N'), # benefício, anúncio, malefício
      #(r'.*cia$', 'N'), # essência, coerência, decência
      #(r'.*ismo', 'N'), # ativismo, machismo
      (r'.*vel', 'ADJ'), # contestável, amável, afável 
      (r'.*cil', 'ADJ'), # fácil, difícil, dócil
      #(r'.*', 'N')
    ]
```

```
tagger_regex = nltk.RegexpTagger(regex, backoff = tagger_dft)
print(tagger_regex.evaluate(sents_teste))
```

```
tagger_unigrama = nltk.UnigramTagger(sents_treino, backoff = tagger_regex)
print(tagger_unigrama.evaluate(sents_teste))
```

```
tagger_bigrama = nltk.BigramTagger(sents_treino, backoff = tagger_unigrama)
print(tagger_bigrama.evaluate(sents_teste))
```

```
 tagger_bigrama_cut = nltk.BigramTagger(sents_treino, cutoff=2, backoff = tagger_unigrama)
 print(tagger_bigrama_cut.evaluate(sents_teste)) 
```

```
from cogroo_interface import Cogroo
cogroo = Cogroo.Instance()
```

```
classes_cogroo=['n', 'prop', 'art', 'pron', 'pron-pers', 'pron-det',
                'pron-indp', 'adj', 'n-adj', 'v', 'v-fin', 'v-inf',
                'v-pcp', 'v-ger', 'num', 'prp', 'adj', 'conj',
                'conj-s','conj-c', 'intj', 'adv', 'xxx']

def _pos_tags(self):
    pos = {}
    pos.update({"n": "substantivo"})
    pos.update({"prop": "nome próprio"})
    pos.update({"art": "artigo"})
    pos.update({"pron": "pronome"})
    pos.update({"pron-pers": "pronome pessoal"})
    pos.update({"pron-det": "pronome determinativo"})
    pos.update({"pron-indp": "substantivo/pron-indp"})
    pos.update({"adj": "adjetivo"})
    pos.update({"n-adj": "substantivo/adjetivo"})
    pos.update({"v": "verbo"})
    pos.update({"v-fin": "verbo finitivo"})
    pos.update({"v-inf": "verbo infinitivo"})
    pos.update({"v-pcp": "verbo particípio"})
    pos.update({"v-ger": "verbo gerúndio"})
    pos.update({"num": "numeral"})
    pos.update({"prp": "preposição"})
    pos.update({"adj": "adjetivo"})
    pos.update({"conj": "conjunção"})
    pos.update({"conj-s": "conjunção subordinativa"})
    pos.update({"conj-c": "conjunção coordenativa"})
    pos.update({"intj": "interjeição"})
    pos.update({"adv": "advérbio"})
    pos.update({"xxx": "outro"})
    return pos

# pos: "part of speech"
pos = cogroo.pos_tags
```

```
macm_words = nltk.corpus.mac_morpho.words()
macm_tagd_words=nltk.corpus.mac_morpho.tagged_words()

macm_tagd_sents = nltk.corpus.mac_morpho.tagged_sents()
macm_sents = nltk.corpus.mac_morpho.sents()


# define tamanho de treino e teste

size_sts_tr = int(len(macm_tagd_sents)*0.9)
size_sts_tst = int(len(macm_tagd_sents)*0.1)

#size_w_tr = int(len(macm_tagd_words)*0.9)
#size_w_tst = int(len(macm_tagd_words)*0.1)

sents_treino_cogroo = macm_tagd_sents[:size_sts_tr]
sents_teste_cogroo = macm_tagd_sents[size_sts_tr:]
```

```
doc = cogroo.analyze(str(sents_treino_cogroo))

sents_positivo = []

for item in doc.sentences:
    sents_positivo = sents_positivo + item.tokens

tagged_positivo = []

for tokens in sents_positivo:
    aux = re.split(r'#',str(tokens))
    aux2 = re.split(r' ',aux[1])
    if(aux2[0] in classes_cogroo):
        tagged_positivo = tagged_positivo + [(aux[0],aux2[0],aux2[1])]

print(tagged_positivo, end='')
```
